```python
import pandas as pd

# Carga el dataset de trabajo
df = pd.read_csv(r"datos originales\netflix.csv")
df
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
df.query("type == 'Movie' & release_year == 2021")["title"]
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
df.query("duration == 90")[["title", "duration"]]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
df.query("country == 'India' & release_year > 2018")[["title"]]
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
df.query("rating == 'TV-14' | rating == 'R'")[["title"]]
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
df.query("director == 'Christopher Nolan' | director == 'Martin Scorsese'")[["title"]]

#https://www.geeksforgeeks.org/pandas-groupby-count-occurrences-in-column/

category = df.query("director == 'Christopher Nolan' | director == 'Martin Scorsese'")[["director"]]

occur = category.groupby(['director']).size() 

occur

```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
df.query("country == 'United States' | country == 'United Kingdom' | country == 'Canada'")[["title"]]

#https://www.geeksforgeeks.org/pandas-groupby-count-occurrences-in-column/

category = df.query("director == 'Christopher Nolan' | director == 'Martin Scorsese'")[["director"]]

occur = category.groupby(['director']).size() 

occur
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
df.query("release_year >= 2015 & release_year <= 2020")[["title"]]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
# df.query("'IMDB Score' >= 6.5 & 'IMDB Score' <= 8.0 & duration == NaN")[["title"]] # Intente este metodo pro no habia manera

df[(df['IMDB Score'] >= 6.5) & (df['IMDB Score'] <= 8.0) & (df['duration'].isna())][["title"]] # https://www.geeksforgeeks.org/filter-pandas-dataframe-with-multiple-conditions/
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
# Voy a usar el mismo metodo en el ejercicio 8 para no estar usando siempre dataquery

df[(df['duration'] == '90 min') | (df['duration'] == '150 min')][["title"]] 
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
df_love = df[(df['title'].str.contains("love")) & (df['type'] == 'Movie')][["title"]] # Con TV Show no habia resultados, y he probado con "Movie para asegurarme que era correcto el filtro"
df_love

```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
df_johnson = df[(df['director'].str.contains("Johnson", regex = True, na = False, case = False))][["title"]] # Con TV Show no habia resultados, y he probado con "Movie para asegurarme que era correcto el filtro"
df_johnson
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python
texto = 'Mike Tyson'

# Busco documentales que sean o bien dirigidos por MT, o que aparezca entre el cast o que su nombre este en el titulo:

df_mt = df[(df['cast'].str.contains(texto, regex = True, na = False, case = False)) # busca a MT en el cast
           |(df['director'].str.contains(texto, regex = True, na = False, case = False)) # busca a MT como director
           | (df['title'].str.contains(texto, regex = True, na = False, case = False))][["title"]] # busca a MT en el titulo
df_mt
```

#### Ejercicio 17: Encuentra los 10 títulos con las puntuaciones más altas en IMDb.

```python
df_filtro = (df[df['IMDB Score'].notna()]  # pongo notna por que si no me saca los nulos      
    .sort_values(by='IMDB Score', ascending=False)  # tengo que decirle que ascending FALSE lo que equivale a DESCENSING de SQL
    .head(10)    # esto es para que me muestre los 10 primeros                        
)
df_filtro['title']
```

#### Ejercicio 18: Encuentra los 5 títulos con las puntuaciones más bajas en IMDb.

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.

```python
df_filtro = df[(df['country'].notna())
           |(df['IMDB Score'].notna())
           | (df['title'].notna())][["title"]] 
df_filtro
```

#### Ejercicio 21: Identifica si hay títulos duplicados en la columna 'title'.

```python
df_title_duplicados = df.duplicated(subset=['title']).sum()

if df_title_duplicados == 0:
    print(f"El dataframe no contiene duplicados")
else:
    print(f"El dataframe contiene {df_title_duplicados} titulos duplicados")
```
